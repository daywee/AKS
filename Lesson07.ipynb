```python
import string
import pandas as pd
import numpy as np
```

# Utils

```python
root_path = './data/Lesson07/similarity/'
labels = ['000', '010', '020', '030', '040', '050', '060', '070', '080', '090']

def read_dataset(label):
    with open(f'{root_path}{label}.txt') as f:
        text = f.read()
        text = text.translate(str.maketrans('', '', string.punctuation))
        text = text.lower()
        
        return text
    
datasets = [read_dataset(label) for label in labels]
```

```python
def create_dictionary(text):
    words = text.split(' ')
    dictionary = set(words)
    
    I = ''
    for word in words:
        I += word
        
        if I in dictionary:
            continue
        
        dictionary.add(I)
        I = word
        
    return dictionary
```

```python
def FCD(text1, text2):
    x, y = set(create_dictionary(text1)), set(create_dictionary(text2))
    
    return (len(x) - len(x&y)) / len(x)
```

# Results

```python
num = len(datasets)
m = np.zeros((num, num))

for i in range(num):
    for j in range(num):
        m[i,j] = round(FCD(datasets[i], datasets[j]), 2)
```

```python
df=pd.DataFrame(m, columns=labels, index=labels)
df
```
